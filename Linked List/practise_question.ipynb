```python
class Node:
    def __init__(self, value):
        self.data = value
        self.next = None
class LinkedList:
    def __init__(self):
        self.head = None
        self.n = 0

    def __len__(self):
        return self.n

    def __str__(self):

        curr = self.head

        result = ''
        while curr is not None:
            result = result + str(curr.data) + ' -> '
            curr = curr.next

        return result[:-4]

    def insert_head(self,value):
        new_node = Node(value)
        new_node.next = self.head

        self.head = new_node
        
        self.n += 1

    def append(self,value):
        new_node = Node(value)
        if self.head is None:
          self.head = new_node
          self.n = self.n + 1
          return
            
        curr = self.head

        while curr.next is not None:
          curr = curr.next
    
        curr.next = new_node
        self.n += 1

    def insert_after(self,after,value):
        new_node = Node(value)
        curr = self.head

        while curr is not None:
            if curr.data == after:
                break
            curr = curr.next

        if curr is not None:
            new_node.next = curr.next
            curr.next = new_node
            self.n += 1
        else:
            print("Item not Found")

    def clear(self):
        self.head = None
        self.n = 0

    def delete_head(self):
        if self.head is not None:
            self.head = self.head.next
            self.n -= 1
        else:
            return "Empty Link List"

    def pop(self):
        if self.head is None:
            return "Empty Linked List"
            
        curr = self.head
        if curr.next is None:
            return self.delete_head()
        
        while curr.next.next is not None:
            curr = curr.next
        curr.next = None
        self.n -= 1

    def remove(self,value):
        if self.head is None:
            return "Empty Linked List"

        if self.head.data == value:
            return self.delete_head()
            
        curr = self.head
        while curr.next is not None:
            if curr.next.data == value:
                break
            curr = curr.next

        if curr.next is None:
            return "Item Not Found"
        else:
            curr.next = curr.next.next
            self.n -= 1

    def search(self,value):
        curr = self.head
        pos = 0

        while curr is not None:
            if curr.data == value:
                return pos
            curr = curr.next
            pos += 1
        return "Item not Found"

    def __getitem__(self,index):
        curr = self.head
        pos = 0

        while curr is not None:
            if pos == index:
                return curr.data
            curr = curr.next
            pos += 1

        return "Index Error"
```

1. Write a python program to find the maximum value in a linked list and replace it with a given value.
   (Assume that the linked list is populated with whole numbers and there is only one maximum value in the linked list)

```python
class ReplaceMax(LinkedList):
    def replace_max(self,num):
        temp = self.head
        max = temp
        while temp is not None:
            if temp.data > max.data:
                max = temp
            temp = temp.next
        max.data = num

L = ReplaceMax()
L.append(0)
L.append(1)
L.append(2)
L.append(3)
L.append(4)
L.replace_max(100)
print(L)

        
```

2. Find the sum of Odd Position Elements of Linked List.

```python
class OddSum(LinkedList):
    def oddSum(self):
        pos = 0
        sum = 0
        curr = self.head

        while curr is not None:
            if pos % 2 != 0:
                sum += curr.data
            curr = curr.next
            pos += 1
        return sum

L = OddSum()
L.append(0)
L.append(1)
L.append(2)
L.append(3)
L.append(4)
L.append(5)
print(L)
print(L.oddSum())

```

3. Write a program which reverses a Linked List containing the integer data(Note: You can not use a new Data Structure)

```python
class Reverse(LinkedList):
    def reverseList(self):
        prev_node = None
        curr_node = self.head
        while curr_node is not None:
            next_node = curr_node.next
            curr_node.next = prev_node
            prev_node = curr_node
            curr_node = next_node
        self.head = prev_node 

L = Reverse()
L.append(0)
L.append(1)
L.append(2)
L.append(3)
L.append(4)
L.append(5)
print(L)
L.reverseList()
print()
print(L)
            
```

```python
class RemoveCharacters(LinkedList):
    def removeCharacters(self):
        temp = self.head

        while temp is not None:
            if temp.data == "*" or temp.data == "/":
                temp.data = " "
                if temp.next.data == "*" or temp.next.data == "/":
                    temp.next.next.data = temp.next.next.data.upper()
                    temp.next = temp.next.next
            temp = temp.next
```
